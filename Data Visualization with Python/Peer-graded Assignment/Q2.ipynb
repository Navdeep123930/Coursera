```
import pandas as pd
import numpy as np
file_path = 'https://cocl.us/datascience_survey_data'
df=pd.read_csv(file_path,index_col=0)
```

```
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```
# Sorting the values based on Very interested
df.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
# Taking the percentage of the responses and rounding it to 2 decimal places 
df= round((df/2233)*100,2)
df.head()
```

```
ax = df.plot(kind='bar', 
                figsize=(20, 8),
                rot=90,color = ['#5cb85c','#5bc0de','#d9534f'],
                width=.8,
                fontsize=14)
ax.set_title('Respondents Percentage Interest in Data Science Areas',fontsize=16)
ax.set_facecolor('white')
ax.legend(fontsize=14,facecolor = 'white') 
ax.get_yaxis().set_visible(False)
# Creating a function to display the percentage.
for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )
plt.show()
```
